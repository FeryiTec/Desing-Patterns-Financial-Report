```python
#A
# Instalar las librerías necesarias si no están instaladas
!pip install pandas numpy

from google.colab import files
import pandas as pd

# Subir el archivo CSV
uploaded = files.upload()

# Cargar el archivo CSV usando el nombre exacto de tu archivo csv
data = pd.read_csv('country_vaccinations.csv')

# Mostrar las primeras filas del DataFrame
data.head()
```

```python
#B
# Mostrar la estructura y tipos de datos
data.info()

# Asegurarse de que las columnas de fecha sean del tipo datetime
data['date'] = pd.to_datetime(data['date'])
data.info()
```

```python
#C
# Separar y contar las vacunas aplicadas por compañía
vaccines_pc= data.groupby('vaccines')['total_vaccinations'].sum()
vaccines_pc
```

```python
#D
# Total de vacunas aplicadas en el mundo
total_vaccinesw = data['total_vaccinations'].sum()
total_vaccinesw
```

```python
#E
# Promedio de vacunas aplicadas por país
average_vaccines = data.groupby('country')['total_vaccinations'].mean()
average_vaccines
```

```python
#F
# Filtrar por fecha
vaccines_jan_29 = data[data['date'] == '2021-01-29']['daily_vaccinations'].sum()
vaccines_jan_29
```

```python
#G
# Crear la nueva columna de diferencias
data['diferencias'] = data['daily_vaccinations'] - data['daily_vaccinations_raw']
conDiferencias = data.copy()
conDiferencias
```

```python
#H
# Obtener fechas más recientes y antiguas
max_date = data['date'].max()
min_date = data['date'].min()
time_period = max_date - min_date
time_period
```

```python
#I
# Crear la columna canVac separando por el carácter ','
data['canVac'] = data['vaccines'].str.split(',').apply(len)
conCantidad = data.copy()
conCantidad
```

```python
#J
antes2020 = data[data['date'] < '2020-12-20']
antes2020.head()
```

```python
#K
pfizer = data[data['vaccines'].str.contains('Pfizer', na=False)]
pfizer.head()
```

```python
#L
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes2020.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
